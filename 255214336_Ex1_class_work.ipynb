```python


import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.simplefilter("ignore", DeprecationWarning)

# create a numpy array of 5 integers
arr = np.array([10, 20, 30, 40, 50])
# normalize the array
arr_stand = (arr - arr.mean()) / arr.std()
print(arr_stand)
```

```python
# create StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# normalize numpy array
scaler.fit_transform(arr.reshape(-1, 1))
# normalize numpy array
scaler.fit_transform(arr.reshape(-1, 1))
```

```python
# create minmax scaler
from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler()

#normalize numpy arr using minmax scaler
min_max.fit_transform(arr.reshape(-1, 1))
```

```python
# Create a dataset with 5 features
# import pandas and its dataframe
import pandas as pd
from pandas import DataFrame
data= {'name':['Rex', 'Peter', 'Sam', 'Rita', 'Hilda'],
'age': [1, 95, 22, 0, 60],
'height':[20, 180, 160, 145,125],
'weight': [0, 100, 75, 55, 40],
'bp': [80, 100, 65, 70, 40]}
#Load data into pandas data frame
frame = DataFrame(data)
print(frame)
#Write dataframe to file
frame.to_csv('patient_data.csv', index=False)
# read file contents back into dataframe
df = pd.read_csv('patient_data.csv')
print(df)
#Replace all zeroes with mean of each column
df = df.replace(0, df.mean(numeric_only = True))
print(df)
#Select all columns except name and create a dataframe
frame2 = df[['age', 'bp', 'height', 'weight']]
print(frame2)
#Apply StandardScaler
scaler.fit(frame2)
# normalize data
scaled = scaler.transform(frame2)
print(scaled)

# print the shape of normalized data
print(scaled.shape)
# print mean and standard deviation values
print("mean : %s " % scaled.mean(axis=0))
print("standard deviation : %s " % scaled.std(axis=0))
#Apply MinMax Scaler
# create another dataframe with all features except name
frame3 = df[['age', 'bp', 'height', 'weight']]
# fit it
print(min_max.fit(frame3))
# normalize it
scaled2 = min_max.transform(frame3)
print(scaled2)
# display mean and standard deviation values
print("mean : %s " % scaled2.mean(axis=0))
print("standard deviation : %s " % scaled2.std(axis=0))
#Select only useful features
from sklearn.feature_selection import VarianceThreshold
# create VarianceThreshold model
sel = VarianceThreshold(threshold=(.5 * (1 - .9)))
#Display preprocessed data based on useful features
useful_features1 = sel.fit_transform(scaled)
print('Useful feature values based on StandardScaler\n')
print(useful_features1)
useful_features2 = sel.fit_transform(scaled2)
print ('\nUseful feature values based on MinMaxScaler')
print(useful_features2)
```
